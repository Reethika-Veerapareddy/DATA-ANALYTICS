```python
#decriptive analysis example
import statistics
import matplotlib.pyplot as plt
mem_names=["anisha","keerthi","laxmi","jyothi","bhumi"]
ages=[22,25,24,23,26]
workout_hours=[2,1.5,2.5,1,3]
calories_burned=[500,250,600,300,750]
print("fitness center dataset")

for i in range(len(mem_names)):
  print(mem_names[i],ages[i],workout_hours[i],calories_burned[i])
avg_workout=sum(workout_hours)/len(workout_hours)
print("average workout hours:",avg_workout)

max_calories=max(calories_burned)
print("maximum calories burned:",max_calories)

min_calories=min(calories_burned)
print("minimum calories burned:",min_calories)

median_calories=statistics.median(calories_burned)
print("median calories burned:",median_calories)

std_dev=statistics.stdev(calories_burned)
print("standard deviation of calories burned:",std_dev)
index=calories_burned.index(max_calories)
print("most active mem",mem_names[index])
plt.bar(mem_names,calories_burned)
plt.xlabel("members")
plt.ylabel("calories burned")
plt.title("Fitness analysis")
plt.show()
plt.pie(calories_burned,labels=mem_names)
plt.title("Fitness analysis")
plt.show()
plt.scatter(mem_names,calories_burned)
plt.xlabel("members")
plt.ylabel("calories burned")
plt.title("Fitness analysis")
plt.show()
```

```python
max_calories=df['Calories_Burned'].max()
print("maximum calories burned:",max_calories)
```

```python

min_calories=df['Calories_Burned'].min()
print("minimum calories burned:",min_calories)
```

```python
avg_workout=df['Session_Duration (hours)'].mean()
print("average workout hours:",avg_workout)
```

```python
index = df['Calories_Burned'].idxmax()
print("most active mem:", index)
```

```python
index = df['Calories_Burned'].idxmin()
print("least active mem:", index)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Calories_Burned'].dropna(), bins=30, edgecolor='black')
plt.title('Distribution of Calories Burned')
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
scatter_data = df[['Calories_Burned', 'Age']].dropna()
plt.scatter(scatter_data['Calories_Burned'], scatter_data['Age'])
plt.title('Calories Burned vs Age')
plt.xlabel('Calories Burned')
plt.ylabel('Age')
plt.grid(axis='both', alpha=0.75)
plt.show()
```
