```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
water_consumption=[10,16,15,19,20,50,70]
days=[2,3,1,6,9,10,12]
print("Dataset")
for i in range(len(water_consumption)):
    print(water_consumption[i]," ",days[i])
print("Average of water consumption")
avg_water_consumption=sum(water_consumption)/len(water_consumption)
print(avg_water_consumption)
water_usage_inday=sum(water_consumption)/sum(days)
print("Water usage in a day")
print( water_usage_inday)
# Calculate the sum of daily increased water consumption
sum_dailyincreased_water = 0

for i in range(1, len(water_consumption)):
    increase = water_consumption[i] - water_consumption[i-1]

    # Only add positive increases
    if increase > 0:
        sum_dailyincreased_water += increase

print("\nSum of daily increased water consumption")
print(sum_dailyincreased_water)
prediction=sum_dailyincreased_water+avg_water_consumption
print("Prediction of water consumption")
print(prediction)
if sum_dailyincreased_water>avg_water_consumption:
    print("shortage will occur")
else:
    print("shortage will not occur")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=days, y=water_consumption)
plt.xlabel('Day')
plt.ylabel('Water Consumption')
plt.title('Water Consumption Over Days')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
temperature = df['meantemp']

# Average temperature
avg_temp = temperature.mean()

print("\nAverage Temperature")
print(avg_temp)

# Calculate positive daily increases
daily_increase = []

for i in range(1, len(temperature)):
    increase = temperature[i] - temperature[i-1]

    if increase > 0:
        daily_increase.append(increase)

# Average increase
avg_increase = sum(daily_increase) / len(daily_increase)

print("\nAverage Daily Increase")
print(avg_increase)

# Last temperature value
last_value = df['meantemp'].iloc[-1]

print("\nLast Temperature")
print(last_value)

# Prediction
prediction = last_value + avg_increase

print("\nPredicted Temperature")
print(prediction)

# Condition
if prediction > avg_temp:
    print("\nWeather temperature may increase")
else:
    print("\nWeather temperature will remain normal")
```
